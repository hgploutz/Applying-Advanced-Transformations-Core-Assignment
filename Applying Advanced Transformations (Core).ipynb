## Cleaning the Data

### Info Dataframe

```python
#Splitting the Hero column at the |
info[['Hero','Publisher']] = info['Hero|Publisher'].str.split('|',expand=True)
```

```python
info.head()
```

#### Adjust Hero|Publisher Column

```python
#Dropping the Hero|Published column since we no longer need it
info = info.drop(columns=['Hero|Publisher'])
info.head()
```

```python
#Viewing the information for the superhero info dataframe
info.info()
```

```python
#Create variable that houses columns that are object types
data_types = info.dtypes
object_data_types_filter = data_types[(data_types == "object")]
object_data_types_filter
```

```python
#For loop to loop through the index of the object columsn and display the unique values in each
for column in object_data_types_filter.index:
    print(column)
    print(info[column].unique())
    print('\n')
```

#### Adjust Measurements Column 

Working on the Height/Weight Column Next

```python
## use .str.replace to replace all single quotes
info['Measurements'] = info['Measurements'].str.replace("'",'"')
## Apply the json.loads to the full column
info['Measurements'] = info['Measurements'].apply(json.loads)
info['Measurements'].head()
```

```python
#Checking to ensure that column is now properly a dictionary
test_measure = info.loc[0, 'Measurements']
print(type(test_measure))
test_measure
```

```python
#Use pd.Series to convert the Measurement column into 2 seperate columns "Height" and "Weight"
height_weight = info['Measurements'].apply(pd.Series)
height_weight
```

```python
#Concatenate those two columns to the original DF and then drop Measurements column
info = pd.concat((info, height_weight), axis = 1)
info.head()
```

```python
#Dropping Measurement column
info = info.drop(columns=['Measurements'])
```

```python
info.head()
```

```python
info.info()
```

```python
#Now to ensure Height and Weight are numeric, remove the " cm" and " kg" from the numbers
info['Height'] = info['Height'].str.replace(" cm",'')
info['Weight'] = info['Weight'].str.replace(" kg",'')
```

```python
#convert Height and Weight to floats
info['Height'] = info['Height'].astype(float)
info['Weight'] = info['Weight'].astype(float)
```

```python
info.head()
```

```python
#Confirming both Height and Weight are now floats
info.info()
```

### Powers DataFrame

#### One Hot Encode Powers Column

```python
powers.info()
```

```python
powers.head()
```

```python
power_test = powers.loc[2,'Powers']
print(type(power_test))
```

```python
#convert string to list
powers['powers_split'] = powers['Powers'].str.split(',')
```

```python
#view the first couple rows to ensure they are lists
powers['powers_split'].head()
```

```python
#get the value counts to check and see how they look 
powers['powers_split'].value_counts()
```

```python
## exploding the column of lists
exploded = powers.explode('powers_split')
exploded[['hero_names','Powers','powers_split']].head(5)
```

```python
## saving the unique values from the exploded column
cols_to_make = exploded['powers_split'].dropna().unique()
cols_to_make
```

```python
#For loop to one hot encode the columns with True/False values for each super ability
for col in cols_to_make:
    powers[col] = powers['Powers'].str.contains(col)
powers.head()
```

```python
#dropping the powers_split and Powers column since they have been One Hot encoded
powers = powers.drop(columns=['Powers','powers_split'])
powers.head()
```

### Merging Powers and Info Dataframes

```python
#Change "hero_names" to "Hero" so I can preform a pd.merge on the two df's
powers.rename(columns={'hero_names': 'Hero'}, inplace=True)
powers.head()
```

```python
#Use pd.merge to combine the two dataframes based on the "Hero" column
merged_df = pd.merge(info,powers, on='Hero')
merged_df.head(2)
```

#### Compare the average weight of super powers who have Super Speed to those who do not.

```python
# Filter the DataFrame based on Super Speed column
has_super_speed = merged_df[merged_df['Super Speed'] == True]
no_super_speed = merged_df[merged_df['Super Speed'] == False]

# Calculate the average weight for each group
average_weight_has_speed = has_super_speed['Weight'].mean()
average_weight_no_speed = no_super_speed['Weight'].mean()

# Compare the average weights
print("Average weight of superpowers with Super Speed:", average_weight_has_speed)
print("Average weight of superpowers without Super Speed:", average_weight_no_speed)
```

- Those superheros who have super speed tend to have a higher average weight then those who do not have super speed

#### What is the average height of heroes for each publisher?

```python
# Group the DataFrame by "Publisher" and calculate the average height
average_height_by_publisher = merged_df.groupby('Publisher')['Height'].mean()

# Display the resulting DataFrame
print(average_height_by_publisher)
```

- Listed above is the average height for each of the Publishers in the merged dataframe
